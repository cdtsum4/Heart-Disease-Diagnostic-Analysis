```python
#Converting Numerical Data into Categorical Data

def heart_disease(row):
    if row == 0:
        return 'Absence'
    elif row == 1:
        return 'Presence'
```

```python
#Applying converted data into our dataset with new column - Heart_Disease

df['Heart_Disease'] = df['target'].apply(heart_disease)
df.head()
```

```python
df.groupby('Heart_Disease')['target'].count()
```

```python
#Converting Numerical Data into Categorical Data

def gender(row):
    if row == 1:
        return 'Male'
    elif row == 0:
        return 'Female'
```

```python
#Applying converted data into our dataset with new column - sex1

df['Gender'] = df['sex'].apply(gender)
df.head()
```

```python
#Converting Numerical Data into Categorical Data

def age_range(row):
    if row>=29 and row<40:
        return 'Young'
    elif row>=40 and row<55:
        return 'Middle Age'
    elif row>55:
        return 'Old'
```

```python
#Applying converted data into our dataset with new column - Age_Range

df['Age_Range'] = df['age'].apply(age_range)
df.head()
```

```python
if 'index' in df.columns:
    df.set_index('index', inplace=True)
else:
    # Add a new index column if it doesn't exist
    df['index'] = range(1, len(df) + 1)
    df.set_index('index', inplace=True)
df   
```
